```python
df['Age_Group']=pd.cut(
    df['Age(year)'],
    bins=[20,35,50,65,90],
labels=['Young','Middle','Senior','Elderly']
                       )
```

```python
df.Age_Group.unique()
```

```python
df.groupby('Age_Group')['Hypertension'].value_counts()
```

```python
ct = pd.crosstab(df['Age_Group'], df['Hypertension'])
print(f'The raw counts are \n {ct} ')
# axis=1 means sum across columns (each row sums to 100%)
# This gives % WITHIN each age group
ct_pct = ct.div(ct.sum(axis=1), axis=0) * 100

print(f'\n Percentage: \n {ct_pct.round(1)}')

sns.heatmap(ct_pct, annot=True, fmt='.1f', cmap='Reds')
plt.title('% Hypertension Stage by Age Group')
plt.show()
```
